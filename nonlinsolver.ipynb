```python
import pandas as pd

import ogstools as ot
```

```python
files = ["MC_ref_newton.log", "MC_ref_trdc.log", "MC_ref_tr.log"]
records = {}
df_records = {}
df_log = {}
df_ts_it = {}
for file in files:
    records[file] = ot.logparser.parse_file(file)
    df_records[file] = pd.DataFrame(records[file])
    df_log[file] = ot.logparser.fill_ogs_context(df_records[file])
    df_ts_it[file] = ot.logparser.time_step_vs_iterations(df_log[file])
```

```python
for file in files:
    df_ts_it[file].plot()
```
